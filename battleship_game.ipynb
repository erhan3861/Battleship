```python
from turtle import *
from random import randint,choice
import winsound
import time

wn = Screen()
wn.title("Battleship Game")
wn.bgcolor("black")
wn.setup(900, 600)
wn.delay(0)

# resimleri pencereye yüklüyoruz
wn.addshape("ship.gif")
wn.addshape("logo.gif")
wn.addshape("start.gif")
wn.addshape("emoji_ship.gif")
wn.addshape("cross.gif")

for i in range(1, 7):
    wn.addshape("ex" + str(i) + ".gif")
    
    
explosion = Turtle() # parlaklık kuklası tanımladık PARANTEZE DİKKAT !!!
explosion.shape("ex1.gif")
explosion.ht() 
explosion.is_active = False

def explode(missile): # Patlama efekti veriliyor. Hangi missile patlıyor ise orada patlama efekti oluşuyor.
    explosion.up()
    explosion.speed(0)
    explosion.goto(missile.pos()) # brightness kuklası missile konumuna git diyoruz.
    explosion.showturtle()
        
    for i in range(1,7):
        time.sleep(0.1)
        explosion.shape("ex" + str(i) + ".gif") # i her döngüde artıyor bu yüzden 10-i ise azalıyor.
        
    explosion.ht()

ship = Turtle() # geminin kodları
ship.shape("ship.gif") # geminin şekli veriliyor
ship.up()
ship.goto(0, 0) # orta aşağı

logo = Turtle()
logo.shape("logo.gif")
logo.up()
logo.goto(-200, 100) # sol yukarı

start = Turtle()
start.shape("start.gif")
start.up()
start.goto(100, -100) # sağ yukarı

emoji_ship = Turtle()
emoji_ship.shape("emoji_ship.gif")
emoji_ship.up()
emoji_ship.ht()

cross = Turtle()
cross.shape("cross.gif")
cross.up()
cross.ht()
cross.goto(-180, 180)

finder = Turtle()
finder.ht()
finder.up()

kutular = []

for i in range(25):
    kutular.append(cross.pos())     
    cross.fd(90)
    if cross.xcor() == 270:
        cross.setx(-180) 
        cross.sety(cross.ycor() - 90)

kutular2 = kutular.copy()
cross.ht()    
# cross.goto(kutular[5])
print(kutular)

ship1 = randint(0,24)
ship2 = randint(0,24)
ship3 = randint(0,24)

while ship1 == ship2 or ship1 == ship3 or ship2 == ship3:
    ship1 = randint(0,24)
    ship2 = randint(0,24)
    ship3 = randint(0,24)

print(ship1)
print(ship2)
print(ship3)
ship_list = [ship1, ship2, ship3]

game_start = False

def start_game(x,y): 
    global game_start
    game_start = True
    ship.ht()
    logo.ht()
    start.ht()
    start.onclick(None)
    wn.bgpic("board_battlegame.gif")
    wn.onscreenclick(move)
    
def stamp(t):
    t.showturtle()
    t.stamp()
    t.ht()
    
def draw_tex(text, color):
    finder.color(color)
    finder.penup()
    finder.goto(0,250)
    finder.clear()
    finder.write(text, align = "center", font = ("Courier", 30, "bold"))

    
def move(x, y):
    global game_start
    
    if game_start:
        game_start = False
        return
    
    finder.goto(x,y)

    for i in kutular:
        if finder.distance(i) < 50:
            finder.goto(i)
            if kutular2.index(i) in ship_list:
                winsound.PlaySound('patlama.wav', winsound.SND_ASYNC)
                time.sleep(0.5)
                emoji_ship.goto(i)
                explode(emoji_ship)
                stamp(emoji_ship)
                ship_list.remove(kutular2.index(i))
            else:
                cross.goto(i)
                stamp(cross)
            kutular.remove(i)
            break
            
    if len(ship_list) == 0:
        draw_tex("You Win  :)  !!!", "spring green")
        wn.onscreenclick(None)
        winsound.PlaySound('oyunbitti.wav', winsound.SND_ASYNC)
        return

    if len(kutular) == 10: # eğer açılmamış 10 kutu kalırsa oyunu bitir
        draw_tex("You Lost :( !!!", "red")
        wn.onscreenclick(None)
        winsound.PlaySound('oyunbitti.wav', winsound.SND_ASYNC)
        return
    
    can = len(kutular) - 10
    draw_tex("Kalan hakkınız = " + str(can), "green")
    
    if can<10:
        draw_tex("Kalan hakkınız = " + str(can), "red")
        
    print(f"x {x} , y = {y}")
    
    
draw_tex("BATTLESHIP GAME", "light blue")   
start.onclick(start_game)
wn.onscreenclick(None)

wn.mainloop()
```
